Wybierz jeden plik Python i:

```python
# risk_score.py after AI REFACTOR
def calculate_risk_score(age: int, cholesterol: int, max_heart_rate: int) -> float:
    """
    Calculate a cardiovascular risk score based on age, cholesterol, and max heart rate.

    Parameters:
        age (int): Age of the person (must be >= 0).
        cholesterol (int): Cholesterol level (must be >= 0).
        max_heart_rate (int): Maximum heart rate (must be > 0).

    Returns:
        float: The calculated risk score, rounded to two decimal places.

    Raises:
        ValueError: If any parameter is negative or if max_heart_rate is not positive.
    """
    if age < 0:
        raise ValueError("Age must be non-negative.")
    if cholesterol < 0:
        raise ValueError("Cholesterol must be non-negative.")
    if max_heart_rate <= 0:
        raise ValueError("Max heart rate must be positive.")

    score = (age * 0.2) + (cholesterol * 0.05) - (max_heart_rate * 0.03)
    return round(score, 2)
```

```python
import pandas as pd
import os

def load_data(path: str) -> pd.DataFrame:
    # check is file exists, validation
    if not os.path.isfile(path):
        raise FileNotFoundError(f"File not found: {path}")
    return pd.read_csv(path)

def basic_report(df: pd.DataFrame) -> dict:
    return {
        'rows': len(df),
        'columns': len(df.columns),
        'missing_values': int(df.isna().sum().sum()),
        'target_distribution': df['target'].value_counts().to_dict() if 'target' in df.columns else {}
    }

if __name__ == '__main__':
    data = load_data("..\\Data\\Heart_Data.csv")
    print(basic_report(data))
```

skrypt generuje bardzo podstawowy raport z:

ilosc kolumn, ilosc rekordow, ilosc brakujacych wartosci (null), rozmieszczenie

```python
# AI EXPANDED
def basic_report_ai(df: pd.DataFrame) -> dict:
    report = {
        'rows': len(df),
        'columns': len(df.columns),
        'missing_values': int(df.isna().sum().sum()),
        'target_distribution': df['target'].value_counts().to_dict() if 'target' in df.columns else {},
        'dtypes': df.dtypes.astype(str).to_dict(),
        'unique_values': df.nunique().to_dict(),
        'numerical_stats': {},
        'missing_rows': int(df.isna().any(axis=1).sum()),
        'columns_with_missing': df.isna().sum()[df.isna().sum() > 0].to_dict(),
        'head': df.head().to_dict(orient='records'),
    }

    # Basic stats for numerical columns
    num_cols = df.select_dtypes(include='number').columns
    for col in num_cols:
        report['numerical_stats'][col] = {
            'mean': df[col].mean(),
            'median': df[col].median(),
            'min': df[col].min(),
            'max': df[col].max(),
        }

    # Optional: Value counts for categorical/small-unique columns
    cat_counts = {}
    for col in df.columns:
        if df[col].nunique() <= 10 or df[col].dtype == 'object':
            cat_counts[col] = df[col].value_counts(dropna=False).to_dict()
    report['categorical_value_counts'] = cat_counts

    return report
```

```python
if __name__ == '__main__':
    data = load_data("..\\Data\\Heart_Data.csv")
    print(basic_report_ai(data))
```
